```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
```

```python
from sklearn.datasets import fetch_california_housing
dataset = fetch_california_housing()
dataset2 = datasets.fetch_california_housing()
```

```python
# Prepare the data

# The target variable is the median house value for California districts,
# expressed in hundreds of thousands of dollars ($100,000).
y = dataset2['target'][:600]
y1 = dataset2['target']

# Use only AveRooms :- average number of rooms per household
# row is instances ':' mean All instances
# [2] feature RM list of 2
X = dataset2['data'][:600,[2]]
X2 = dataset2['data'][:,[1]]
```

```python
#Visualise the data
plt.scatter(X,y)
plt.xlabel('AveRooms')
plt.ylabel('Price')
```

```python
print(X2[0])
```

```python
#Visualise the data
plt.scatter(X2,y1)
plt.xlabel('HouseAge')
plt.ylabel('Price')
```

```python
# choose model and train model
# Initiate a linear model
model = linear_model.LinearRegression()

# Train the model by use fit() to find appropriate linear
model.fit(X,y) #calling a learning algorithm
```

```python
print('Price','=',model.coef_,'* RM +',model.intercept_)
```

```python
# visualise linear
y_pred = model.coef_ * X + model.intercept_
```

```python
#plot data
plt.scatter(X,y,color='black')
plt.scatter(X,y_pred,color='blue',linewidth=3)
plt.xlabel('RM')
plt.ylabel('Price')
```

```python
# Make a prediction
print(model.coef_*7+model.intercept_)
```

```python
print(model.predict([[7]])) #when input into predict func. should be 2d array
```
